```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

Datasets = []
PREDICTORS = ["PwmD", "PwmE"]   
TARGET = ["Theta", "X", "Y"]       
SETS = [
    "Train_1",
    "Train_2",
    "Test_1",
    "Test_2",
    "Test_3",
    "Val",
    "Lsg_1",
    "Lsg_2",
]


for i in range(4):
    Dataset = pd.read_excel(f"../../RotedData/Data.xlsx", f"D{i+1}")   
    # Criar as colunas de deltas automaticamente
    for var in TARGET:
        Dataset[f"Delta{var}"] = Dataset[var].shift(-1) - Dataset[var]    
           
    Datasets.append(Dataset)

for i in range(4):   
    Dataset = pd.read_csv(f"../../Data/Data{i + 1}.csv")
        
    # Criar as colunas de deltas automaticamente
    for var in TARGET:
        Dataset[f"Delta{var}"] = Dataset[var].shift(-1) - Dataset[var]    
    
    Datasets.append(Dataset)
    

for i in range(len(Datasets)):
    print(f"++++++++++++++++++++ Dataset {i+1} +++++++++++++++++++++++")
    print(Dataset.head(5))
    
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 4, figsize=(18, 10))
axes = axes.flatten()

for i, Dataset in enumerate(Datasets):
    
    ax = axes[i]
    
    ax.plot(
        Dataset["X"], 
        Dataset["Y"], 
        linewidth=1
    )
    
    ax.set_title(SETS[i])
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.grid(True)

plt.tight_layout()
plt.show()
```
